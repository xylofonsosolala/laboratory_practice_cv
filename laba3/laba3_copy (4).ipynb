Читаем все кадры и выводим информацию

```python
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print(f"Не удалось открыть видео: {video_path}")
else:
    print(f"Видео открыто успешно!")

i = 1
success = True
imgs = []
print(f"Начало чтения кадров:")

while success:
    print(f"Чтение кадра {i}...", end=" ")
    success, img = cap.read()

    if success and img is not None:
        print(f"Успешно, размер: {img.shape}")
        img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        imgs.append(img_grey)
        i = i + 1
    else:
        print(f"Конец видео или ошибка чтения")
        break

cap.release()
print(f"\n Всего прочитано кадров: {len(imgs)}")

if imgs:
    print(f"   Размер кадра: {imgs[0].shape}")

    plt.figure(figsize=(15, 5))
    for idx, gray_img in enumerate(imgs[:5]):
        plt.subplot(1, 5, idx+1)
        plt.imshow(gray_img, cmap='gray')
        plt.title(f'Кадр {idx+1}')
        plt.axis('off')
    plt.tight_layout()
    plt.show()

    print(f"\nИнформация о кадрах:")
    for idx, gray_img in enumerate(imgs):
        print(f"   Кадр {idx+1}: мин={gray_img.min()}, макс={gray_img.max()}, среднее={gray_img.mean():.2f}")
else:
    print("Нет кадров для отображения")
```

Задача нашей лабы определить движение мышки на видео и построить траекторию ее движения. Поскольку мышка движется в постоянном фоне, наилучшим решением данной задачи будет являться метод построения модели фона. Для начала реализуем самый простой метод. А именно метод усреднения фона. Данный метод в реальных условиях не возможен (ну или мало возможен), поскольку для него необходимо усреднить пиксели всех кадров, но в нашем примере такая информация нам сразу доступна. Для начала посмотрим функцию сложения изображений попиксельно.

```python
def sr_mod_fon (arr_image):
  mod_fon = np.zeros(arr_image[1].shape, dtype='uint8')
  sum=np.zeros([480,640])
  for k in range(len(arr_image)-10):
    image=arr_image[k]
    for i in range(len(image)):
      for j in range(len(image[i])):
        sum[i,j]=sum[i,j]+image[i,j]
    print(k)

  sum=sum/k
  return sum
```

Данная функция складывает значения пикселей, которые занимают одно и тоже пространственное расположение. А после прохода по всему массиву берется среднее от количества изображений.

```python
amt=sr_mod_fon(imgs)
amt
```

Поскольку на выходе мы получаем нецелые значения пикселей, то их нужно округлить. Для это воспользуемся функцией библиотеки np для перевода в требуемый формат изображения, а именно np.uint8

```python
new=np.uint8(amt)
```

```python
imgs[2005]
```

```python
diff_image = cv2.absdiff(imgs[2005], new)
ret,thresh = cv2.threshold(diff_image, 20, 255,cv2.THRESH_BINARY)
thresh=cv2.morphologyEx(thresh, cv2.MORPH_OPEN, np.ones((5,5),np.uint8), iterations=1)
thresh=cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, np.ones((5,5),np.uint8), iterations=1)
thresh
```

```python
mu=cv2.moments(thresh)
mc = (mu['m10'] / (mu['m00'] + 1e-5), mu['m01'] / (mu['m00'] + 1e-5))
mc
```

```python
image_test=cv2.copyTo(imgs[2005],imgs[2005])
cv2.circle(image_test,(int(mc[0]), int(mc[1])), 5, (0,0,255), -1)
```

```python
def mod_fon (arr_image):
  mod_fon = np.zeros(arr_image[1].shape, dtype='uint8')
  mom=[None]*len(arr_image)
  for k in range(len(arr_image)-10):
    mod_fon=cv2.addWeighted(arr_image[k],0.5,mod_fon,0.5,0)
    diff_image=cv2.addWeighted(arr_image[k],1,mod_fon,-1,0)
    ret,thresh = cv2.threshold(diff_image, 20, 255,cv2.THRESH_BINARY)
    mu=cv2.moments(thresh)
    mom[k]=(mu['m10'] / (mu['m00'] + 1e-5), mu['m01'] / (mu['m00'] + 1e-5))
    if(mom[k][0]==0 and mom[k][1]==0):
      mom[k]=mom[k-1]
    if(k!=0):
      if((mom[k][0]-mom[k-1][0])>30 or (mom[k][1]-mom[k-1][1])>30):
        mom[k]=mom[k-1]
  return mom
```
